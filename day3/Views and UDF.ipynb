```sql
create or replace view dev.aryan.driver_nationality as 
(select nationality, count(*) as count from dev.aryan.drivers group by nationality order by count)
```

```sql
create or replace function dev.aryan.full_name(first_name string, last_name string)
returns string
return concat(first_name,' ' ,last_name)
```

```sql
select dev.aryan.full_name("Aryan", "Sinha") as name 
```

```sql
select *, dev.aryan.full_name(forename,surname) as full_name from dev.aryan.drivers
```

```sql
CREATE OR REPLACE FUNCTION dev.aryan.mask_email(email STRING)
RETURNS STRING
RETURN CONCAT(
  '*******',
  SUBSTRING(email, INSTR(email, '.com'))
);
```

```sql
select dev.aryan.mask_email('aryan.sinha@databricks.com') as masked_email
```
