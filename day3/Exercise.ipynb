```python
df=spark.read.csv("/Volumes/dev/naval/raw/task/financial_dataset.csv",header=True,inferSchema=True)
```

```python
df.write.mode("overwrite").saveAsTable("dev.aryan.financial_metadata")
```

```python
%sql
create or replace view dev.aryan.financial_summary as 
(select TransactionType, AmountUSD, count(*) as count from dev.aryan.financial_metadata group by TransactionType, AmountUSD order by count desc)
```

```python
%sql
USE CATALOG dev;
USE SCHEMA aryan;
```

```python
%sql
CREATE OR REPLACE FUNCTION dev.aryan.usd_to_inr(amount DOUBLE)
RETURNS DOUBLE
RETURN amount * 93;
```
